```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"C:\Users\Admin\Downloads\IRIS (1).csv")
print("Feature types:")
print(dataset.dtypes)
```

```python
dataset.hist(figsize=(10, 8))
plt.suptitle('Histograms of IRIS Features')
plt.tight_layout()
plt.show()
```

```python
dataset.plot(kind='box', subplots=True, layout=(2,2), figsize=(10, 8), sharex=False, sharey=False)
plt.suptitle('Boxplots of IRIS Features')
plt.tight_layout()
plt.show()
```

```python
print("\nOutliers based on IQR method:")
for column in dataset.select_dtypes(include=['float64', 'int64']):
    Q1 = dataset[column].quantile(0.25)
    Q3 = dataset[column].quantile(0.75)
    IQR = Q3 - Q1
    outliers = dataset[(dataset[column] < Q1 - 1.5 * IQR) | (dataset[column] > Q3 + 1.5 * IQR)]
    print(f"{column}: {len(outliers)} outliers")
```
