# Data Analytics I: 
- Show linear regression technique for user values.
- Create a Linear Regression Model using Python/R to predict salary of 15 years of experience using salary_csv file.


```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
dataset = pd.read_csv("Salary_data.csv")
dataset.head()
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
sns.heatmap(dataset.corr())
```

```python
X = dataset.drop(columns={'Salary'})

Y = dataset['Salary']

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state = 5)
```

```python
print("X_train.shape : ",X_train.shape)
print("Y_train.shape : ",Y_train.shape)
print("X_test.shape : ",X_test.shape)
print("Y_test.shape : ",Y_test.shape)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Mean Squared Error (MSE):", mean_squared_error(y_pred,Y_test))
```

```python
print("Mean Absolute Error (MAE):", mean_absolute_error(y_pred,Y_test))
```

```python
error = np.abs(Y_test-y_pred)

plt.scatter(Y_test,y_pred,c=error)
plt.plot([min(Y_test),max(Y_test)],[min(Y_test),max(Y_test)] , color = 'red' , linestyle ='--')
plt.title("Actual V/S predicted salary")
plt.xlabel("Actual Salary")
plt.ylabel("Predicted Salary")
plt.colorbar()
plt.show()
```
